## A contour plot can be createad with the plt.contour function

It takes three arguments: a grid of x values, a grid of y values, and a grid of z values

```python
f = lambda x,y: np.sin(x)*np.cos(y)
x = np.linspace(-5,5,50)
y = np.linspace(-5,5,40)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)

plt.contour(X,Y,Z,cmap="copper")
None
```

## Topic 6 : Area of a region

Find the area below $f(x) = -x^{2} + 4x + 3$ and above $g(x) = -x^{3} + 7x^{2} - 10x + 5 $ over the interval $1 < x < 2$

```python
from sympy import *

def f(t):
    return -t**2 + 4*t + 3
def g(t):
    return -t**3 + 7*t**2 - 10*t + 5

t = Symbol("t")
def find_area(f,g,t,a,b):
    a = Integral(f-g,(t,a,b)).doit()
    return a

find_area(f(t),g(t),t,1,2)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np

t = np.linspace(0, 3, 100)
plt.figure(figsize = [8, 6])
plt.plot(t, -t**2 + 4*t + 3, 'r-')
plt.plot(t, -t**3+7*t**2 - 10*t + 5, 'b-')
plt.axvline(x = 1)
plt.axvline(x = 2)
plt.title("Area between the curves")
plt.legend(['$-t^2 + 4^t + 3$','$-t^3+7t^2 - 10^t + 5$'])

p = np.linspace(1, 2, 50)
plt.fill_between(p, -p**2 + 4*p + 3, -p**3+7*p**2 - 10*p + 5)
plt.show()
```
